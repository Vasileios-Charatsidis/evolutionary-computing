```python
import numpy as np
import matplotlib.pyplot as plt
```

## Differential evolution plot

```python
def plot(file):
    poptest = []
    crosstest = []
    difftest = []
    pertest = []
    
    populationSize = []
    crossoverRate = []
    differentialRate = []
    pertubationSize = []
    Score = []
    Runtime = []

    function = file.readline()
    function = function.replace('\n','')
    
    for line in file:
        myline = line.replace('\n', '')
        spl = myline.split(':')
        if spl[0] == "populationSize":
            poptest.append(int(spl[1]))
        if spl[0] == "crossoverRate":
            crosstest.append(float(spl[1]))
        if spl[0] == "differentialRate":
            difftest.append(float(spl[1]))
        if spl[0] == "pertubationSize":
            pertest.append(int(spl[1]))
        if spl[0] == "Score":
            Score.append(float(spl[1]))
            populationSize.append(poptest[-1])
            crossoverRate.append(crosstest[-1])
            differentialRate.append(difftest[-1])
            pertubationSize.append(pertest[-1])
    
    fig = plt.figure()
    plt.plot(populationSize, Score, '.')
    plt.title("differential evolution: populationsize dependence on " + function)
    plt.xlabel("populationsize")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/dif_ev/'+ function + '_population.png')
    
    fig = plt.figure()
    plt.plot(crossoverRate, Score, '.')
    plt.title("differential evolution: crossoverrate dependence on " + function)
    plt.xlabel("crossoverRate")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/dif_ev/' + function + '_crossover.png')
    
    fig = plt.figure()
    plt.plot(differentialRate, Score, '.')
    plt.title("differential evolution: differentialrate dependence on " + function)
    plt.xlabel("differentialrate")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/dif_ev/' + function + '_differential.png')
    
    fig = plt.figure()
    plt.plot(pertubationSize, Score, '.')
    plt.title("differential evolution: pertubationSize dependence on " + function)
    plt.xlabel("pertubationSize")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/dif_ev/' + function + '_pertubation.png')


cigar_dif = open('BentCigar2.txt', 'r')
schaffer_dif = open('Schaffers2.txt', 'r')
katsuura_dif = open('Katsuura2.txt', 'r')
sphere_dif = open('Sphere2.txt', 'r')

plot(sphere_dif)
plot(katsuura_dif)
plot(cigar_dif)
plot(schaffer_dif)
```

## Particle swamb plot

```python
def plot(file):
    
    function = file.readline()
    function = function.replace('\n','')
    
    curtest = []
    inditest = []
    poptest = []
    velotest = []
    boundtest = []
    frictest = []
    
    CurrentWeight = []
    IndividualWeight = []
    populationSize = []
    velocityMax = []
    boundMethod = []
    friction = []
    Score = []

    for line in file:
        myline = line.replace('\n', '')
        spl = myline.split(':')
        if spl[0] == "weightCurrent":
            curtest.append(float(spl[1]))
        if spl[0] == "weightIndividual":
            inditest.append(float(spl[1]))
        if spl[0] == "populationSize":
            poptest.append(int(spl[1]))
        if spl[0] == "velocityMax":
            velotest.append(float(spl[1]))
        if spl[0] == "boundMethod":
            boundtest.append(str(spl[1]))
        if spl[0] == "friction":
            frictest.append(float(spl[1]))
        if spl[0] == "Score":
            Score.append(float(spl[1]))
            CurrentWeight.append(curtest[-1])
            IndividualWeight.append(inditest[-1])
            populationSize.append(poptest[-1])
            velocityMax.append(velotest[-1])
            boundMethod.append(boundtest[-1])
            friction.append(frictest[-1])
    
    fig = plt.figure()
    plt.plot(CurrentWeight, Score, '.')
    plt.title("partical swarm: currentWeight dependence on " + function)
    plt.xlabel("CurrentWeight")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/par_sw/'+ function + '_CurrentWeight.png')
    
    fig = plt.figure()
    plt.plot(IndividualWeight, Score, '.')
    plt.title("partical swarm: IndividualWeight dependence on " + function)
    plt.xlabel("IndividualWeight")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/par_sw/' + function + '_IndividualWeight.png')
    
    fig = plt.figure()
    plt.plot(populationSize, Score, '.')
    plt.title("partical swarm: populationSize dependence on " + function)
    plt.xlabel("populationSize")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/par_sw/' + function + '_populationSize.png')
    
    fig = plt.figure()
    plt.plot(velocityMax, Score, '.')
    plt.title("partical swarm: velocityMax dependence on " + function)
    plt.xlabel("velocityMax")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/par_sw/' + function + '_velocityMax.png')
    
    fig = plt.figure()
    plt.plot(friction, Score, '.')
    plt.title("partical swarm: friction dependence on " + function)
    plt.xlabel("friction")
    plt.ylabel("Score")
    plt.legend()
    plt.show()
    fig.savefig('./Data/par_sw/' + function + '_friction.png')


cigar_par = open('BentCigar1.txt', 'r')
schaffer_par = open('Schaffers1.txt', 'r')
katsuura_par = open('Katsuura1.txt', 'r')
sphere_par = open('Sphere1.txt', 'r')

plot(schaffer_par)
plot(cigar_par)
plot(katsuura_par)
plot(sphere_par)
```
